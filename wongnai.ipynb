# Import Library

```python
import pandas as pd
import pymysql
from sqlalchemy import create_engine
import folium
from folium.plugins import HeatMap
```

# Prepare Data

```python
res = pd.read_sql("SELECT * FROM w_restaurant",db)
cat = pd.read_sql("SELECT * FROM w_category",db)
region = pd.read_sql("SELECT * FROM w_region",db)
```

# Heat Map

```python
def make_heatmap(cat_id,name):
    restype = res[res['category_id'] ==cat_id]

    hmap = folium.Map(location=[13.736717, 100.523186], zoom_start=7)
    hm_wide = HeatMap( list(zip(restype.latitude.values, restype.longitude.values)),
                       min_opacity=0.1,
                       radius=10, blur=10, 
                       max_zoom=2, 
                       max_val=3
                    )
    hmap.add_child(hm_wide)
    hmap.save('heatmap_shop_of_'+name+'.html')

    hmap2 = folium.Map(location=[13.736717, 100.523186], zoom_start=7)
    hm_wide2 = HeatMap( list(zip(restype.latitude.values, restype.longitude.values,restype.number_of_bookmarks.values)),
                       min_opacity=0.1,
                       radius=10, blur=10, 
                       max_zoom=1, 
                       max_val=max(restype.number_of_bookmarks.values)/2
                    )
    hmap2.add_child(hm_wide2)
    hmap2.save('heatmap_bookmark_of_'+name+'.html')

    hmap3 = folium.Map(location=[13.736717, 100.523186], zoom_start=7)
    hm_wide3 = HeatMap( list(zip(restype.latitude.values, restype.longitude.values,
                                 restype.number_of_checkins.values + restype.number_of_been_here.values)),
                       min_opacity=0.1,
                       radius=10,
                       blur=10, 
                       max_zoom=1, 
                       max_val= max(restype.number_of_checkins.values + restype.number_of_been_here.values)/2
                    )
    hmap3.add_child(hm_wide3)
    hmap3.save('heatmap_reach_of_'+name+'.html')
```

# Region Analysis

```python
def process_data_of_cat(cat_id):
    temp = res[res['category_id'] ==cat_id]
    temp["number_of_shop"] = 1
    temp = temp.groupby(["region_id"]).sum()[["number_of_bookmarks","number_of_checkins","number_of_been_here","number_of_shop"]]
    temp["number_of_reach"] = temp["number_of_checkins"] +  temp["number_of_been_here"] 
    temp = temp.drop(columns=["number_of_checkins","number_of_been_here"],axis=1)
    temp["reach_per_shop"] =  (temp["number_of_reach"]-temp["number_of_shop"])/temp["number_of_shop"]
    top_worst = temp[temp["number_of_reach"]>0].sort_values(by=['reach_per_shop'],ascending=True)[:5]
    top_best = temp[temp["number_of_reach"]>0].sort_values(by=['reach_per_shop'],ascending=False)[:5]
    return top_worst,top_best 

def add_region(pd):
    pd["region"] = ""
    for i in range(len(pd.index)):
        pd["region"][pd.index[i]] = region_table[region_table["id"] == pd.index[i]]["name"].iloc[0]
    return pd
```

```python
def cat_process(cat_id):
    name = cat[cat["id"]==cat_id].iloc[0]["name"]
    x = name.replace("/","")
    make_heatmap(cat_id,x)
    raw = process_data_of_cat(cat_id)
    add_region(raw[0]).to_csv("top_worst_of_"+x,sep="\t")
    add_region(raw[1]).to_csv("top_best_of_"+x,sep="\t")
```
